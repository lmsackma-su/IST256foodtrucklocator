```python
def nearest_food_truck(city): 
    params = {'city': city} 
    response = requests.get('http://data.streetfoodapp.com/1.1/schedule/%s' % city,params=params)
    if response.ok:
        schedule_data = response.json()
        return schedule_data
    else:
        print("Sorry! That isn't a city we have a food truck vendor for. Please enter another city.")
        print(response.text)
```

```python
#getting location of someone
def get_coordinates(search):
    url = 'https://nominatim.openstreetmap.org/search'  # base URL without paramters after the "?"
    options = { 'q' : search, 'format' : 'json'}
    response = requests.get(url, params = options)            
    geodata = response.json()
    coords = {'lat' : float(geodata[0]['lat']), 'lng' : float(geodata[0]['lon'])}
    return coords
```

```python
#finding closest vendors near user
def foodnearme(search):
    coords = get_coords(search)
    #opening list of vendors 
    lst = []
    #storing nearest food trucks into a variable
    g = nearest_food_truck('boston')
    for store in g['vendors']:
        storelat = g['vendors'][store]['last']['latitude']
        storelon = g['vendors'][store]['last']['longitude']
        mylat=coords[0]
        mylon=coords[1]
        if storelat<=mylat + 0.03 and storelat>= mylat - 0.03:
            if storelon<=mylon + 0.03 and storelon>=mylon - 0.03:
                 lst.append(store)       
    return lst
```
